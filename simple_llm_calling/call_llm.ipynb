```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()

llm = ChatGoogleGenerativeAI(model="gemma-4-31b-it")

response = llm.invoke("How many big cities does Germany have?")

print(response.text)
```

```python
response = llm.invoke([
    ["system", "You are a helpful assistant that answers in short one line"],
    ["human","How many big cities does Germany have?"]
])
print(response.text)
```

```python
prompt = "Write a single tagline for a german coffee shop. Only output the tagline, nothing else."
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)

response = llm.invoke(prompt)
print(response.text)
```

##Call LLM Using Groq Models

```python
from langchain_groq import ChatGroq

llm = ChatGroq(model="qwen/qwen3-32b", reasoning_format="parsed")

messages = [
    (
        "system",
        "You always give short answer to a question"
    ),
    (
        "human",
        "How many big cities does Germany have?"
    )
]
response = llm.invoke(messages)
print(response.content)
```
