```python
import chromadb

client = chromadb.Client()
```

```python
collection = client.create_collection("news_articles")
```

```python
collection.add(
    ids=["id1", "id2", "id3","id4"],
    documents=[
        "Apple is leading in smart phone game with Iphone sales up by 35%",
        "Tesla booked a minor profit of 1 billion $ in Q2",
        "Apples are high in fiber, vitamin C, and various antioxidants",
        "SpaceX got NASA contract worth 10 billion $",
    ]
)
```

```python
# Fetch document + embeddings
data = collection.get(
    include=["documents", "embeddings", "metadatas"]
)

# print embeddings
for i, emb in enumerate(data["embeddings"]):
    print(f"\nDocument {i+1}")
    print(f"Text: {data["documents"][i]}")
    print(f"Embedding length: {len(emb)}")
    print(f"First 10 values: {emb[:10]}")
```

```python
results = collection.query(
    query_texts=["This is a query related to Elon Musk"],
    n_results=2
)

print(results)
```
